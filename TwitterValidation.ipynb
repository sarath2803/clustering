```python
df.drop(['id','location'],axis=1,inplace=True)
```

```python
df
```

```python
df=df.loc[df['target']!='Irrelevant']
df
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
df['target']=df['target'].map({'Neutral':0,'Positive':1,'Negative':-1})
df
```

```python
tweets=df.text
tweets
```

```python
tweets=tweets.str.replace("[^a-zA-Z0-9]+"," ")
tweets
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer=SnowballStemmer('english')
tk=TweetTokenizer()
```

```python
tweets=tweets.apply(lambda x:[stemmer.stem(token.lower()) for token in tk.tokenize(x)]).apply(lambda token:" ".join(token))
tweets
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
sw=stopwords.words('english')
sw
```

```python
tweets=tweets.apply(lambda x:[token for token in tk.tokenize(x) if token not in sw]).apply(lambda token:" ".join(token))
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer()
X=vect.fit_transform(tweets)
print(X)
```

```python
vect.get_feature_names_out()
```

```python
y=df['target']
y
```

```python
#RF
#KNN
#AB
#SVC
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.svm import SVC
rf=RandomForestClassifier()
knn=KNeighborsClassifier()
ab=AdaBoostClassifier()
svc=SVC()
```

```python
li=[rf,knn,ab,svc]
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
for i in li:
  print("model:",i)
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(classification_report(y_test,y_pred))
```
